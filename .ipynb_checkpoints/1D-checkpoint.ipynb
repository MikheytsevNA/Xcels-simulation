```python
l = subprocess.check_output(['ls',], cwd = name+ r'/BasicOutput/data/bz2dxy')
files = l.split()
files = [i.decode('ascii') for i in files]
#print(files)
```

```python
def reparce(files, name, field):
    bz = []
    for file in files:
        with gzip.open(name +r"/BasicOutput/data/" + field + "/" + file, mode = "rb") as f:
            line_of_data = np.frombuffer(f.read(), dtype = np.float32)
            #by_field_2_i = np.reshape(line_of_data, shape_2dfield)/RelativisticField
            bz.append(line_of_data/RelativisticField)
    return bz
```

```python
# репарс полей
by,bz,bx = [],[],[]
bz = np.array(reparce(files[:-2], name, "bz2dxy"))
by = np.array(reparce(files[:-2], name, "by2dxy"))
#bx = np.array(reparce(files[:-2], name, "bx2dxy"))

ey,ez,ex = [],[],[]
ey = np.array(reparce(files[:-2], name, "ey2dxy"))
ez = np.array(reparce(files[:-2], name, "ez2dxy"))
#ex = np.array(reparce(files[:-2], name, "ex2dxy"))
```

```python
# разделение на падающее и отраженное поля
by_r = (ez+by)/2
by_i = (-ez+by)/2

bz_r = (-ey+bz)/2
bz_i = (ey+bz)/2
```

## Зависимость эффективность от концентрации

```python
rep = reparse()
```

```python
folder_name = "./data/1D_en_data"
```

```python
n_0, lambda_s, energy, num_n = rep.energy(folder_name, "n0*.bin")

for i in range(len(n_0)):
    lambda_s[i] = [x for _,x in sorted(zip(n_0[i],lambda_s[i]))]
    energy[i] = [x for _,x in sorted(zip(n_0[i],energy[i]))]
    n_0[i].sort()
```

```python
def built_graph_compare(r1, limit):
    range4 = np.zeros(num_n)
    for i in np.arange(0, num_n):
        mask1 = np.where((lambda_s.T[i] > r1[0]) & (lambda_s.T[i] < r1[1]))[0]
        for i1 in mask1:
            range4[i] += energy[i1,i]*100
    fig = plt.figure(figsize=(20,10))
    #plt.plot(n_0_circular_steep[1],range1, label = "$\circlearrowright$ steep".format(r1[0], r1[1]), linewidth = 3)
    #plt.plot(n_0_circular_long[1],range2, label = "$\circlearrowright$ long".format(r1[0], r1[1]), linewidth = 3)
    #plt.plot(n_0_linear_steep[1],range3, label = r"$\updownarrow$ steep".format(r1[0], r1[1]), linestyle = "none", marker = ".", markersize = 10)
    plt.plot(n_0[1],range4, label = r"$\updownarrow$ long".format(r1[0], r1[1]), linestyle = "none", marker = ".", markersize = 20)
    plt.xlabel(r"$n_0$")
    plt.ylabel(r"$\eta$ ({} - {} $\mu m$), %".format(r1[0], r1[1]))
    plt.xticks((np.arange(0,40,2)))
    #plt.yticks(np.arange(0,limit, 0.2))
    plt.ylim(1e-3, limit)
    plt.xlim(1, 40)
    plt.grid()
    plt.legend()
```

### Эффективность в двух диапазонах - от 3 до 4 микрон, более 4 микрон.

```python
interactive_plot = interactive(built_graph_compare,
                            r1 = widgets.FloatRangeSlider(min=0.02,max=50,step=0.1,value= [3.00, 4.0]),
                            limit = widgets.FloatText(value= 2.0))
output = interactive_plot.children[-1]
output.layout.height = '700px'
interactive_plot
```
